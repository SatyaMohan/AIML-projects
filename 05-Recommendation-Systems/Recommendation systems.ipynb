## Treat data sparsity:

Based on the above plots, we can see that the data is very sparse. To make our recommendations more relevant, we need to make it a bit dense so that we don't recommend edge case products (such as products with very few ratings).  

We will try to make data less sparse by only considering the products which have got at least a minimum number of ratings

```python
# Get the product-rating count
prod_rating_count = ratings_df.groupby('productId').count()

sns.distplot(prod_rating_count['rating'])
```

```python
prod_rating_count['rating'].value_counts().sort_index().head(60)
```

**Based on above output and general user behaviour of considering what is reliable, it would be safe to keep the threshold of 10 ratings**

```python
prod_rating_count = prod_rating_count.reset_index()
prod_ratings_df = ratings_df.merge(prod_rating_count[prod_rating_count.rating >= 10]['productId'], on='productId', how='inner')
```

```python
len(prod_ratings_df)
```

We will try to make data less sparse by only considering the users who have given at least a minimum number of ratings


```python
# Get the user-rating count
user_rating_count = prod_ratings_df.groupby('userId').count()

sns.distplot(user_rating_count['rating'])
```

```python
user_rating_count['rating'].value_counts().sort_index().head(60)
```

**In our analysis above, we noted that our data spans across 190 months.
```python
user_rating_count = user_rating_count.reset_index()
final_ratings_df = prod_ratings_df.merge(user_rating_count[user_rating_count.rating >= 50]['userId'], on='userId', how='inner')
```

### 2. User-User collaborative filtering 

In user-user collaborative filtering, we find the similarities between different users and consider that similar users will like similar products. This forms the basis for us to make recommendations hence there is a fair degree of personalization

```python
# We will use surprise library to perform user-user collaborative filtering

from surprise import KNNWithMeans, Dataset, accuracy, Reader, SVD, BaselineOnly
from surprise.model_selection import train_test_split as sptts, cross_validate, GridSearchCV
from sklearn.model_selection import train_test_split as sktts
```

```python
# Transform the dataframe to surprise dataset object
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(final_ratings_df, reader)

# Split the data in 70:30 ratio
trainset, testset = sptts(data, test_size=0.3, random_state=42)
```

```python
# Train the algo with trainset
algo = KNNWithMeans(k=50, sim_options={'name': 'cosine', 'user_based': True})
algo.fit(trainset)
```

```python
# Get predections on test
test_pred = algo.test(testset)
```

```python
test_pred
```

```python
# Calculate RMSE
print("User-based KNN Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
benchmark = []
# Iterate over KNN and SVD algorithms
for algorithm in [SVD(), KNNWithMeans()]:
    # Perform cross validation
    results = cross_validate(algorithm, data, measures=['RMSE'], cv=3, verbose=True, n_jobs=-1)
    
    # Get results & append algorithm name
    tmp = pd.DataFrame.from_dict(results).mean(axis=0)
    tmp = tmp.append(pd.Series([str(algorithm).split(' ')[0].split('.')[-1]], index=['Algorithm']))
    benchmark.append(tmp)
    
pd.DataFrame(benchmark).set_index('Algorithm').sort_values('test_rmse')
```

**We see that *SVD* algorithm is giving the best test_rmse, so we will pick this algorithm and tune the hyper parameters** 

```python
# Keep a test set aside 
train_df, test_df = sktts(final_ratings_df, test_size=0.3, random_state=42)

# Convert to surprise dataset
train_data = Dataset.load_from_df(train_df, reader)

# We will use train_df for hyper-parameter tuning to avoid data-leakage issue
param_grid = {'n_factors':[50,100,150],'n_epochs':[20,30],  'lr_all':[0.005,0.01],'reg_all':[0.02,0.1]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse'], cv=3, joblib_verbose=2, n_jobs=-1)
gs.fit(train_data)
params = gs.best_params['rmse']
svdtuned = SVD(n_factors=params['n_factors'], n_epochs=params['n_epochs'],lr_all=params['lr_all'], reg_all=params['reg_all'])
```

```python
# Use the tuned model
svdtuned.fit(trainset)

# Get predections on test
tuned_test_pred = svdtuned.test(testset)

# Calculate RMSE
print("User-based SVD Model : Test Set")
accuracy.rmse(tuned_test_pred, verbose=True)
```

## Making new recommendations: 

We will make top 5 new recommendations for each user

```python
# For each user, get the list of products which he/she haven't rated
# For each such product predict the rating and store top 5 rated products

user_set = set(final_ratings_df['userId'])
product_set = set(final_ratings_df['productId'])

user_product_summary = list()
for cnt, user in enumerate(user_set):
    rated_products_set = set(final_ratings_df[final_ratings_df.userId==user]['productId'])
    new_products_set = product_set - rated_products_set
    new_product_ratings = dict()
    for new_product in new_products_set:
        new_product_ratings[new_product] = svdtuned.predict(user, new_product).est
    user_product_summary.append({'userId': user, 
                                'productIds': sorted(new_product_ratings, 
                                                              key=lambda x: new_product_ratings[x], 
                                                              reverse=True)[:5]})
    print(f"Done prediction for user({cnt+1}/{len(user_set)}): {user}")
```

```python
# Format and convert to dataframe

for row in user_product_summary:
    row['product1'], \
    row['product2'], \
    row['product3'], \
    row['product4'], \
    row['product5']   = row['productIds']

recommendations_df = pd.DataFrame(data=user_product_summary, columns=['userId', 'product1','product2', 'product3',
                                                                      'product4', 'product5'])
recommendations_df.head()
```

Above are the **top-5 recommendations** (product1 > product5) of **new products** for a few users  
Recommendations for other users can be obtained by appropriately querying the **recommendations_df** dataframe
